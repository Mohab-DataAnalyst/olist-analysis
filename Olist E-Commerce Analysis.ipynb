#### Time & Trend Analysis

```python
# order purchase date range
min_date = orders_df['order_purchase_timestamp'].dt.date.min()
max_date = orders_df['order_purchase_timestamp'].dt.date.max()

print('Orders Date Range: ', min_date, '-', max_date)
```

```python
# revenue over time
orders_order_items_merge = orders_order_items_merge.set_index('order_purchase_timestamp')
# filtering only delivered orders
orders_order_items_merge = orders_order_items_merge[orders_order_items_merge['order_status'] == 'delivered']

# revenue by month
monthly_revenue = (
    orders_order_items_merge['order_item_revenue']
    .resample('MS')
    .sum()
    .reset_index()
    .rename(columns={'order_purchase_timestamp' : 'order_purchase_month', 'order_item_revenue' : 'monthly_revenue'})
)

# 3-month moving avg
monthly_revenue['revenue_ma3'] = monthly_revenue['monthly_revenue'].rolling(window=3).mean()

monthly_revenue.head()
```

```python
# plotting revenue over time
plt.figure(figsize=(12,6))
sns.lineplot(y='monthly_revenue', x='order_purchase_month', data=monthly_revenue, label='Monthly Revenue', color='blue')
sns.lineplot(y='revenue_ma3', x='order_purchase_month', data=monthly_revenue, label='3-Month Moving Average', color='orange')
plt.title('Revenue over Time')
plt.xlabel('Months')
plt.ylabel('Revenue ($M)')
plt.show()
```

```python
# getting the best month
best_revenue_month = monthly_revenue.loc[[monthly_revenue['monthly_revenue'].idxmax()]]
best_revenue_month
```

```python
# revenue monthly seasonality
monthly_revenue['month'] = monthly_revenue['order_purchase_month'].dt.month
revenue_seasonality = (
    monthly_revenue
    .groupby('month')
    .agg({'monthly_revenue' : 'mean'})
    .reset_index()
)
revenue_seasonality['monthly_revenue_k'] = revenue_seasonality['monthly_revenue']/1e3 # scaling to K

revenue_seasonality.head()
```

```python
# plotting revenue monthly seasonality
plt.figure(figsize=(12,6))
sns.barplot(y='monthly_revenue_k', x='month', data=revenue_seasonality)
plt.title('Revenue Seasonality')
plt.xlabel('Months')
plt.ylabel('Revenue ($K)')
plt.show()
```

```python
# getting the best month in seasonality
best_revenue_month_seasonality = revenue_seasonality.loc[[revenue_seasonality['monthly_revenue'].idxmax()]]
best_revenue_month_seasonality
```

```python
# getting the worst month in seasonality
worst_revenue_month_seasonality = revenue_seasonality.loc[[revenue_seasonality['monthly_revenue'].idxmin()]]
worst_revenue_month_seasonality
```

```python
# orders over time
orders_df_delivered = orders_df[orders_df['order_status'] == 'delivered'] # filtering only delivered orders
monthly_orders = (
    orders_df_delivered
    .set_index('order_purchase_timestamp')['order_id']
    .resample('MS')
    .count()
    .reset_index()
    .rename(columns={'order_purchase_timestamp' : 'order_purchase_month', 'order_id' : 'monthly_orders'})
)
monthly_orders.head()
```

```python
# plotting orders over time
plt.figure(figsize=(12,6))
sns.lineplot(y='monthly_orders', x='order_purchase_month', data=monthly_orders)
plt.title('Orders over Time')
plt.xlabel('Months')
plt.ylabel('Orders')
plt.show()
```

```python
# getting the best month
best_order_month = monthly_orders.loc[[monthly_orders['monthly_orders'].idxmax()]]
best_order_month
```

```python
# orders monthly seasonality
monthly_orders['month'] = monthly_orders['order_purchase_month'].dt.month
orders_seasonality = (
    monthly_orders
    .groupby('month')
    .agg({'monthly_orders' : 'mean'})
    .reset_index()
)
orders_seasonality.head()
```

```python
# plotting orders monthly seasonality
plt.figure(figsize=(12,6))
sns.barplot(y='monthly_orders', x='month', data=orders_seasonality)
plt.title('Orders Seasonality')
plt.xlabel('Months')
plt.ylabel('Orders')
plt.show()
```

```python
# getting the best month in seasonality
best_order_month_seasonality = orders_seasonality.loc[[orders_seasonality['monthly_orders'].idxmax()]]
best_order_month_seasonality
```

```python
# getting the worst month in seasonality
worst_order_month_seasonality = orders_seasonality.loc[[orders_seasonality['monthly_orders'].idxmin()]]
worst_order_month_seasonality
```

```python
# The monthly revenue appears highly correlated with the monthly orders amount
# The extreme drops in monthly revenue are justified by the tiny amount of orders that were placed
```

```python
# sellers by state distribution
sellers_by_state = (
    sellers_df
    .groupby('seller_state')
    .agg({'seller_id' : 'count'})
    .sort_values(by='seller_id', ascending=False)
    .reset_index()
    .rename(columns={'seller_id' : 'seller_count'})
)

# adding percentages
sellers_by_state['seller_count_pct'] = ((sellers_by_state['seller_count'] / qty_sellers) * 100).round(1)
sellers_by_state
```

```python
# payment types distribution
payment_types_distrib = (
    payments_df
    .groupby('payment_type')
    .agg({'order_id' : 'count'})
    .sort_values(by='order_id', ascending=False)
    .reset_index()
    .rename(columns={'order_id' : 'payment_count'})
)
# adding percentages
qty_payments = payment_types_distrib['payment_count'].sum()
payment_types_distrib['payment_count_pct'] = ((payment_types_distrib['payment_count'] / qty_payments) * 100).round(1)
payment_types_distrib
```

```python
# reviews distribution
reviews_distrib = (
    reviews_df
    .groupby('review_score')
    .agg({'order_id' : 'count'})
    .sort_values(by='review_score', ascending=True)
    .reset_index()
    .rename(columns={'order_id' : 'reviews'})
)
# adding percentages
qty_reviews = reviews_distrib['reviews'].sum()
reviews_distrib['reviews_pct'] = ((reviews_distrib['reviews'] / qty_reviews) * 100).round(1)
reviews_distrib
```

```python
# plotting reviews distribution
plt.figure(figsize=(12,6))
sns.barplot(y='reviews', x='review_score', data=reviews_distrib)
plt.title('Review Scores Distribution')
plt.xlabel('Score')
plt.ylabel('Reviews')
plt.show()
```

```python
# avg review score by category

reviews_category_merge = ( # merging required dfs
    reviews_df
    .merge(orders_df, on='order_id')
    .merge(order_items_df, on='order_id')
    .merge(products_df, on='product_id')
    .merge(prod_category_trans_df, on='product_category_name')
)

reviews_score_by_category = ( # grouping avg scores by category, top 10
    reviews_category_merge
    .groupby('product_category_name_english')
    .agg({'review_score' : 'mean'})
    .sort_values(by='review_score', ascending=False)
    .reset_index()
    .rename(columns={'review_score' : 'avg_review_score'})
    .head(10)
)
reviews_score_by_category
```

```python
# plotting avg reviews score by category
plt.figure(figsize=(12,6))
sns.barplot(y='product_category_name_english',x='avg_review_score',data=reviews_score_by_category)
plt.title('Average Review Scores by Product Category')
plt.xlabel('Avg Score')
plt.ylabel('Category')
plt.show()
```

```python
# avg review score by state

reviews_states_merge = ( # merging required dfs
    reviews_df
    .merge(orders_df, on='order_id')
    .merge(customers_df, on='customer_id')
)

reviews_score_by_state = ( # grouping avg scores by state, top 10
    reviews_states_merge
    .groupby('customer_state')
    .agg({'review_score' : 'mean'})
    .sort_values(by='review_score', ascending=False)
    .reset_index()
    .rename(columns={'review_score' : 'avg_review_score'})
    .head(10)
)
reviews_score_by_state
```

```python
# plotting avg reviews score by state
plt.figure(figsize=(12,6))
sns.barplot(y='customer_state',x='avg_review_score',data=reviews_score_by_state)
plt.title('Average Review Scores by Customer States')
plt.xlabel('Avg Score')
plt.ylabel('State')
plt.show()
```

```python
# avg review score by seller

reviews_sellers_merge = ( # merging required dfs
    reviews_df
    .merge(orders_df, on='order_id')
    .merge(order_items_df, on='order_id')
    .merge(sellers_df, on='seller_id')
)

reviews_score_by_seller = ( # grouping avg scores by seller, top 10
    reviews_sellers_merge
    .groupby('seller_id')
    .agg({'review_score' : 'mean'})
    .sort_values(by='review_score', ascending=False)
    .reset_index()
    .rename(columns={'review_score' : 'avg_review_score'})
)
reviews_score_by_seller.head(10)
```

#### RFM Analysis

```python
# setting a reference date
reference_date = orders_df['order_purchase_timestamp'].dt.date.max()+dt.timedelta(days=1)
reference_date
```

```python
# rfm building
rfm = (
    customers_order_items_merge
    .groupby('customer_unique_id')
    .agg({
        'order_purchase_timestamp' : lambda x : (reference_date - x.dt.date.max()).days,
        'order_id' : 'count',
        'order_item_revenue' : 'sum'
    })
    .rename(columns={
        'order_purchase_timestamp' : 'recency',
        'order_id' : 'frequency',
        'order_item_revenue' : 'value'
    })
)
rfm.head()
```

```python
# computing rfm quantiles
quantiles = rfm.quantile(q=[0.25,0.50,0.75])
quantiles
```

```python
# 3% of customers have frequency higher or equal to 3 orders
(rfm.query("frequency >= 3").shape[0] / rfm.shape[0]) * 100
```

```python
# computing rfm scores
def rfm_score(x,p,d):
    if p == 'recency':
        if x >= d.loc[0.75, p]:
            return 1
        elif x >= d.loc[0.50, p]:
            return 2
        elif x >= d.loc[0.25, p]:
            return 3
        else:
            return 4
    elif p == 'value':
        if x >= d.loc[0.75, p]:
            return 4
        elif x >= d.loc[0.50, p]:
            return 3
        elif x >= d.loc[0.25, p]:
            return 2
        else:
            return 1
    else: # quantiles will not work due to the extreme skewness in the data
        if x >= 5:
            return 4
        elif x >= 3:
            return 3
        elif x >= 2:
            return 2
        else: 
            return 1

rfm['R'] = rfm['recency'].apply(rfm_score, args=('recency', quantiles))
rfm['F'] = rfm['frequency'].apply(rfm_score, args=('frequency', quantiles))
rfm['M'] = rfm['value'].apply(rfm_score, args=('value', quantiles))

rfm.head()
```

```python
# building rfm segment and total score
rfm['rfm_segment_id'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)
rfm['rfm_score'] = rfm['R'] + rfm['F'] + rfm['M']
rfm.head()
```

```python
# total score quantiles for thresholds
sum_quantiles = rfm['rfm_score'].quantile([0.25,0.50, 0.75])
high_threshold = sum_quantiles[0.75]
low_threshold = sum_quantiles[0.25]
```

```python
# assigning segment labels
def assign_label(score):
    if score >= high_threshold:
        return 'High-Value'
    elif score >=low_threshold:
        return 'Medium-Value'
    else:
        return 'Low-Value'

rfm['rfm_label'] = rfm['rfm_score'].apply(assign_label)

rfm.head()
```

```python
# segment label distribution
rfm = rfm.reset_index()

rfm_label_distrib = (
    rfm
    .groupby('rfm_label')
    .agg({'customer_unique_id' : 'count'})
    .reset_index()
    .rename(columns={'customer_unique_id' : 'customers'})
)
rfm_label_distrib
```

```python
# plotting segment label distribution
order = ['Low-Value', 'Medium-Value', 'High-Value']
plt.figure(figsize=(12,6))
sns.barplot(y='customers', x='rfm_label', data=rfm_label_distrib, order=order)
plt.title('Customer Distribution by RFM Label')
plt.xlabel('Label')
plt.ylabel('Customers Count')
plt.show()
```

```python
# computing rfm quantiles
# f is not dependent on quantiles since most of the customers are one-time purchase ones (see customer behavior analysis)

rfm_quantiles = rfm[['R', 'F', 'M']].quantile([0.25, 0.50, 0.75])
r_low_thresh = rfm_quantiles.loc[0.25, 'R']
r_medium_thresh = rfm_quantiles.loc[0.50, 'R']
r_high_thresh = rfm_quantiles.loc[0.75, 'R']
f_low_thresh = rfm_quantiles.loc[0.25, 'F']
f_medium_thresh = 2
f_high_thresh = 3
m_low_thresh = 1
m_medium_thresh = rfm_quantiles.loc[0.50, 'M']
m_high_thresh = rfm_quantiles.loc[0.75, 'M']
```

```python
# assigning rfm segment
def assign_segment(row):
    
    r = row['R']
    f = row['F']
    m = row['M']
    
    if r >= r_high_thresh and f >= f_high_thresh and m >= m_high_thresh:
        return "Champions"
    elif m >= m_high_thresh:
        return "Whales"
    elif f >= f_high_thresh:
        return "Loyal"
    elif f >= f_high_thresh and m <= m_medium_thresh:
        return "Promising"
    elif r >= r_high_thresh and f <= f_low_thresh:
        return "Rookies"
    elif r <= r_low_thresh and f <= f_low_thresh:
        return "Slipping"
    else:
        return 'Others'

rfm['rfm_segment'] = rfm.apply(assign_segment, axis=1)

rfm.head()
```

```python
# segment distribution
rfm_segment_distrib = (
    rfm
    .groupby('rfm_segment')
    .agg({'customer_unique_id' : 'count'})
    .reset_index()
    .rename(columns={'customer_unique_id' : 'customers'})
)
rfm_segment_distrib
```

```python
# plotting segment distribution
order = ['Others', 'Slipping', 'Rookies', 'Promising', 'Loyal', 'Whales', 'Champions']
plt.figure(figsize=(12,6))
sns.barplot(y='customers', x='rfm_segment', data=rfm_segment_distrib, order=order)
plt.title('Customer Distribution by RFM Segment')
plt.xlabel('Segment')
plt.ylabel('Customers Count')
plt.show()
```

```python
# computing the means of r, f, m scores to make a cluster bar chart
rfm_means = (
    rfm
    .groupby('rfm_segment')
    .agg({
        'R' : 'mean',
        'F' : 'mean',
        'M' : 'mean'
    })
    .reset_index()
)
rfm_means.head()
```

```python
# melting the rfm_means df to make a cluster bar chart
rfm_means_melt = rfm_means.melt(
    id_vars = 'rfm_segment',
    value_vars = ['R', 'F', 'M'],
    var_name = 'Metric',
    value_name = 'Score'
)
rfm_means_melt.head()
```
